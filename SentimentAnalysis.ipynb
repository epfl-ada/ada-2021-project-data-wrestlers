# Projet ADA : Sentiment analysis / Flair

```python
import bz2
import json
import pandas as pd
import string
import re
import nltk
from nltk.stem import WordNetLemmatizer
import matplotlib.pyplot as plt
import flair
import vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

## 1) Data preparation

```python
# Import data

data = []
with bz2.open('data/quotes-2019-nytimes.json.bz2', "rt") as data_file :
    for line in data_file :
        try :
            lignedata = json.loads(line)
            data.append(lignedata)
        except:
            continue

df = pd.DataFrame()

df["quoteID"] = list(map(lambda d : d["quoteID"], data))
df["quotation"] = list(map(lambda d : d["quotation"], data))
df["speaker"] = list(map(lambda d : d["speaker"], data))
df["qids"] = list(map(lambda d : d["qids"], data))
df["date"] = list(map(lambda d : d["date"], data))
df["numOccurrences"] = list(map(lambda d : d["numOccurrences"], data))
df["probas"] = list(map(lambda d : d["probas"], data))
df["urls"] = list(map(lambda d : d["urls"], data))
df["phase"] = list(map(lambda d : d["phase"], data))
```

```python
# Pre-processing

df_select = df.copy(deep=True)

# Dropping rows where the main speaker is "None"
indexNames = df_select[df_select['speaker']=='None'].index
df_select.drop(indexNames, inplace=True)

# Dropping rows where the probability of the main speaker is < 0.5
indexNames = df_select[df_select['probas'].apply(lambda x: x[0][1]).astype(float)<0.5].index
df_select.drop(indexNames, inplace=True)

# Dropping NA quotations 
indexNames = df_select[df_select.quotation.isna()].index
df_select.drop(indexNames, inplace=True)

# Converting the date column into date format
df_select.date = pd.to_datetime(df_select.date)
```

```python
# Construction of a daraframe with quotations about Trump 
# (and speaker is not Trump or someone with Trump in his/her name)
df_about = df_select[df_select.quotation.str.lower().str.contains("trump")].copy(deep=True)
df_about.drop(df_about[df_about.speaker.str.lower().str.contains("trump")].index, inplace=True)
df_about = df_about.reset_index(drop=True)

# Travailler avec un plus petit jeu de donn√©es
df_about_test = df_about[0:5].copy(deep=True)

df_about_test
```

We made the choice to select the quotes about Trump in a quite general way (contains 'trump') so as not to lose too many quotes. We are aware that some of those quotes will not actually be about Donald Trump but we expect that those will not influence the results much.

## 2) Flair

```python
# Flair function for sentiment analysis
def flair_analysis(quotation):
    classifier = flair.models.TextClassifier.load('en-sentiment')
    sentence = flair.data.Sentence(quotation)
    classifier.predict(sentence)
    value = sentence.labels[0].to_dict()['value']
    confidence = sentence.labels[0].to_dict()['confidence']
    return [value, confidence]
```

```python
# Sections of dataframe to run progressively in case of error after a while
size = 2
df_list = [df_about_test[pos:pos+size] for pos in range(0, len(df_about_test), size)]

# Flair sentiment analysis
for n, df_seq in enumerate(df_list):
    df_out = df_seq.copy()
    df_out.loc[:,"Flair"] = df_out.quotation.apply(flair_analysis)
    df_list[n] = df_out
    print("df_list: {} / index_df_about_test: {}". format(n, df_out.index))
```

```python
# Constructing the new dataframe with Flair sentiment analysis
df_about_test_flair = pd.concat(df_list, ignore_index=True)
df_about_test_flair = pd.concat([df_about_test_flair,pd.DataFrame(df_about_test_flair['Flair'].tolist(), columns=['FlairSentiment', 'FlairConfidence'])], axis=1)

df_about_test_flair
```

```python
# Saving dataframe
df_about_test_flair.to_pickle("data/df_about_test_flair.pkl")
```

```python
# Removing row with FlairConfidence < 0.5
indexNames = df_about_test_flair[df_about_test_flair.FlairConfidence.apply(lambda x: x<0.9)].index
df_about_test_flair.drop(indexNames, inplace=True)
df_about_test_flair.reset_index(drop=True, inplace=True)
```

## 3) VaderSentiment

```python
# Sections of dataframe to run progressively in case of error after a while
size = 2
df_list = [df_about_test_flair[pos:pos+size] for pos in range(0, len(df_about_test_flair), size)]

# Analyser
analyzer = SentimentIntensityAnalyzer()

# Vader sentiment analysis
for n, df_seq in enumerate(df_list):
    df_out = df_seq.copy()
    df_out.loc[:,"Vader"] = df_out.quotation.apply(lambda x: analyzer.polarity_scores(x))
    df_list[n] = df_out
    print("df_list: {} / index_df_about_test_flair: {}". format(n, df_out.index))
```

```python
# Adding Vader sentiment analysis in a new dataframe
df_about_test_sentiment = pd.concat(df_list, ignore_index=True)

# Column with Vader compound
df_about_test_sentiment['VaderCompound']  = df_about_test_sentiment.Vader.apply(lambda x: x['compound'])

# Column with final Vader sentiment
df_about_test_sentiment['VaderSentiment'] = df_about_test_sentiment.VaderCompound.apply(lambda x: 'POSITIVE' if x >=0 else 'NEGATIVE')

df_about_test_sentiment
```

```python
# Saving dataframe
df_about_test_sentiment.to_pickle("data/df_about_test_sentiment.pkl")
```
